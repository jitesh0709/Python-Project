```python
def extract_epd(txt):
    data = ""
    collecting = False
    
    for ch in txt:
        if ch == "(":
            collecting = True
            continue
        elif ch == ")":
            break
        if collecting:
            data += ch
            
    return data
```

```python
df["Episodes"]=df["Title"].apply(extract_epd)
```

```python
def extraction_time(txt):
    check=False
    data=""
    for i in range(len(txt)):
        if txt[i] == ')':
            for j in range(i+1,i+20):
                data+=txt[j]

            return data
```
